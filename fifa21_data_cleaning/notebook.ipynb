Now writing a function to saperate players who have been loaned out and those who are Free. Also create a start_data, end date and contract period column, and position them right next to Contract column.

```python

def extract_contract_info(x):
    # Handle 'Free' or 'On Loan' contracts
    if x == "Free" or "On Loan" in x:
        start_date = np.nan
        end_date = np.nan
        contract_length = np.nan
    else:
        split_data = x.split(" ~ ")

        # Check if the data can be split into two parts (start and end dates)
        if len(split_data) != 2:  
            return np.nan, np.nan, np.nan
        
        start_date = split_data[0].strip()  # Remove any leading/trailing spaces
        end_date = split_data[1].strip()

        try:
            contract_length = int(end_date[:4]) - int(start_date[:4])
        except (ValueError, IndexError):
            return np.nan, np.nan, np.nan

    return start_date, end_date, contract_length

 # Apply function to Contract column
new_cols = ['Contract Start Date', 'Contract End Date', 'Contract Length']
df = df_copy["Contract"].apply(lambda x: pd.Series(extract_contract_info(x), index=new_cols))

# Update the position of the new columns besides Contract Columns
for i in range(len(new_cols)):
    df_copy.insert(loc=df_copy.columns.get_loc("Contract")+i+1, column=new_cols[i], value=df[new_cols[i]])

df_copy.head()
```

Now with that, its important to categorize the contract type. We will create a new column called `Contract Type` and categorize the contract type into 3 categories: `Loan`, `Free`, `Contract`.

```python
def categorize_contract_status(x):
    if x == "Free":
        return "Free"
    elif "On Loan" in x:
        return "On Loan"
    else:
        return "Contract"

# Apply function to Contract column and add new column besides Contract length
df_copy.insert(loc=df_copy.columns.get_loc("Contract Length")+1, column="Contract Status", value=df_copy["Contract"].apply(categorize_contract_status))

df_copy.head()
```

### Height & Weight

Players height should be in cm and weight should be in kg. We will convert the height to cm and weight to kg.

```python
# Exploring Height
df_copy['Height'].unique()
```

remove "cm", convert to int and make every value in cm

```python
def combine_height(x):
    if "cm" in x:
        return int(x.strip("cm"))
    else:
        feet , inches = x.split("'")
        total_inches = int(feet) * 12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

# Apply function to Height column
df_copy['Height'] = df_copy['Height'].apply(combine_height)
df_copy["Height"].unique()
    
```

```python
# Rename the Height column name
df_copy.rename(columns={"Height": "Height (cm)"}, inplace=True)
df_copy.head()
```

```python
# Exploring Weight
df_copy['Weight'].unique()
```

Data type is Object. We need to convert lbs into kgs, remove kg and convert to int

```python
def convert_weight(x):
    if "kg" in x:
        return int(x.strip("kg"))
    elif "lbs" in x:
        pounds = int(x.strip("lbs"))
        return round(pounds * 0.45359237)
    else:
        # Handle unexpected formats
        return None
    
# Apply function to Weight column
df_copy['Weight'] = df_copy['Weight'].apply(convert_weight)
df_copy["Weight"].unique()
```

```python
# Rename the Weight column name
df_copy.rename(columns={"Weight": "Weight (kg)"}, inplace=True)
df_copy.head()
```

### Hits Column
No of time a player has been searched on the website. We will fill the missing values with 0

```python
# Exploring Hits variable
df_copy["Hits"].unique()
```

It appears that the `Hits` column has a lot of discripencies. We will fill the missing values with 0, convert the data type to int, remove the K from the values and convert them to int.

```python
# Remove the comma from the Hits column & convert to int

def convert_hits(x):
    if not isinstance(x, str):
        return x
    if "K" in x:
        return int(float(x.strip("K")) * 1000)
    elif "M" in x:
        return int(float(x.strip("M")) * 1000000)
    elif x == "nan":
        return np.nan
    else:
        return int(x)

# Apply function to Hits column
df_copy["Hits"] = df_copy["Hits"].apply(convert_hits)
df_copy["Hits"].unique()
```
